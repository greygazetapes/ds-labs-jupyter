## **3. Уникальность: обнаружение дубликатов через множества**

```python
# Множество уникальных и дублирующихся record_id
seen = set()
dup_record_ids = set()

for r in records:
  rid = r['user_id']
  if rid in seen:
    dup_record_ids.add(rid)
  else:
    seen.add(rid)

seen, dup_record_ids
```

```python
unique_records = []
seen = set()

for r in records:
    rid = r["user_id"]
    if rid not in seen:
        unique_records.append(r)
        seen.add(rid)

unique_records
```

## **4. Минимальная очистка: пропуски и шум**

```python
noise_user_id = [r for r in records if r['user_id'] is not None and not (r['user_id'] >= 0)]
noise_movie_id = [r for r in records if r['movie_id']is not None and not (r['movie_id'] > 0)]
noise_rating = [r for r in records if r['rating'] is not None and not (0 <= r['rating'] <= 10)]
noise_minutes_watched = [r for r in records if r['minutes_watched'] is not None and not (0 <= r['minutes_watched'] <= 200)]
noise_date = [r for r in records if r['date'] is not None and (isinstance(r['date'], str)) and ((not 1 <= int(r['date'].split('-')[1]) <= 12) or not (1 <= int(r['date'].split('-')[2]) <= 31))]
```

```python
clean_records = [r for r in records
                 if r.get('minutes_watched') is not None and 0 <= r.get('minutes_watched') <= 200
                 and r.get('user_id') is not None and (r['user_id'] >= 0)
                 and r.get('rating') is not None and (0 <= r.get('rating') <= 10)
                 and r.get('movie_id') is not None and (r.get('movie_id') > 0)
                 and r.get('date') is not None
                 and (isinstance(r['date'], str))
                 and (1 <= int(r['date'].split('-')[1]) <= 12)
                 and (1 <= int(r['date'].split('-')[2]) <= 31)
]
clean_records
```

## **5. Группировка и агрегирование без Pandas**

```python
count_by_movie_id = defaultdict(int)
for r in clean_records:
  count_by_movie_id[r['movie_id']] += 1

dict(count_by_movie_id)
```

```python
watched_minutes_movie = defaultdict(list)
for r in clean_records:
  watched_minutes_movie[r['movie_id']].append(r['minutes_watched'])

dict(watched_minutes_movie)
```

```python
avg_by_watched_minutes = {movie: mean(vals) for movie, vals in watched_minutes_movie.items()}
avg_by_watched_minutes
```

```python
summary = [
    {"movie": movie, "count": count_by_movie_id[movie], "avg_watched_minutes": round(avg_by_watched_minutes[movie], 2)}
    for movie in count_by_movie_id
]
summary
```

## **6. Сортировка результатов и применение генераторов**

```python
summary_sorted = sorted(summary, key=lambda x: x['avg_watched_minutes'], reverse=True)
summary_sorted
```

```python
def valid_records(source):
    for r in source:
        sc = r.get("minutes_watched")
        if sc is not None and 0 <= sc <= 200:
            yield r

list(valid_records(unique_records))
```

```python
vals = (r["minutes_watched"] for r in valid_records(unique_records))
vals_list = list(vals)
overall_avg = mean(vals_list) if vals_list else None
overall_avg
```

## **Контрольные вопросы**

**Чем отличается список (list) от множества (set) при решении задач контроля уникальности?**

List хранит элементы в порядке добавления, допускает дубликаты.
Set хранит только уникальные элементы, порядок не гарантируется.

**Для чего в аналитических задачах используется структура «список словарей» (records)?**

Это универсальная структура для хранения табличных данных в Python.
Каждый словарь соответствует одной записи (строке таблицы).
Ключи словаря — названия полей (столбцов).

**В каких случаях генераторы предпочтительнее списков?**

Когда нужно ленивое вычисление (элементы создаются по мере запроса).
При работе с большими данными, чтобы не загружать всю коллекцию в память.
В случаях, когда результат используется один раз (например, в цикле или при суммировании).

**Как выполняется группировка данных без Pandas? Какие структуры данных для этого удобны**

Можно использовать стандартные структуры:
* dict + list: ключ — категория, значение — список элементов.
* defaultdict(list) из collections: автоматически создаёт списки для новых ключей.
* Counter или defaultdict(int) для подсчётов.

**Почему сортировка результатов анализа важна для интерпретации данных?**

Упорядочивание помогает увидеть тренды (например, рост по годам).
Делает результаты читаемыми: легче понять, где максимум/минимум.
