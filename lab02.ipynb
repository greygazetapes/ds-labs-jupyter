## **3. Уникальность: обнаружение дубликатов через множества**

```python
# Множество уникальных и дублирующихся record_id
seen = set()
dup_record_ids = set()

for r in records:
  rid = r['user_id']
  if rid in seen:
    dup_record_ids.add(rid)
  else:
    seen.add(rid)

seen, dup_record_ids
```

```python
unique_records = []
seen = set()

for r in records:
    rid = r["user_id"]
    if rid not in seen:
        unique_records.append(r)
        seen.add(rid)

unique_records
```

## **4. Минимальная очистка: пропуски и шум**
